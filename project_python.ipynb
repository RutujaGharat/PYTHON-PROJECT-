```python
# to import beautifulsoup library
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
```

```python
#import library and webdrive

from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from bs4 import BeautifulSoup
import matplotlib.pyplot as plt
```

```python
#create the object of service class to open a  ChromeDriver that matches your installed version of Google Chrome.
service = Service(ChromeDriverManager().install())


driver = webdriver.Chrome(service=service)
```

```python
# to open the site to browser 

driver.get('https://in.puma.com/in/en/womens')
driver.refresh()
```

```python
# to check the code 

soup = BeautifulSoup(driver.page_source,"html.parser")

print(soup.prettify())
```

```python
# to extract product name (div)


title_h3 = soup.find_all('div', attrs={'class' : "flex flex-col w-full"})

title_h3
```

```python
# to create a empty list and extract the title name

titles =[]


for title in title_h3:
    titles.append(title.text)

titles
```

```python
# to extract price 

price_span = soup.find_all('span', attrs={'class': 'whitespace-nowrap text-base text-puma-red font-bold'})

price_span
```

```python
# to create a empty list and extract the price 

prices = []


for price in price_span:
   p =  price.text
   p = p.replace(',','')
   p = p.replace('â¹','')

   prices.append(int(p))



prices
```

```python
#oirgnal price 
og_price  = soup.find_all('span', attrs={'class': "whitespace-nowrap text-base line-through opacity-50 override:text-puma-black-300 override:opacity-100"})

og_price
```

```python
# to extract orignal price 
ogprices = []


for pr in og_price:
   p =  pr.text
   p = p.replace(',','')
   p = p.replace('â¹','')
   ogprices.append(int(p))

ogprices
```

```python
# to extract discount
discount_per   = soup.find_all('span', attrs={'class': "min-w-content tw-1ghqyvy tw-noc9l0 absolute top-0 left-0 z-[5]"})

discount_per
```

```python
discounts =[]


for dis in discount_per:
    discounts.append(dis.text)

discounts
```
